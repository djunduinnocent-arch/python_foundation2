```python
my_list = list(range(-50,50,2))
my_list_b = [10]*len(my_list)
dict_DF ={"a": my_list, "b":my_list_b}
import pandas
df = pandas.DataFrame(dict_DF)

df["t"] = df["a"] / df["b"]
df["t"]
```

```python
import numpy
df["s"] = numpy.cos(df["t"])
df["s"]
```

```python
df["s5"] = df["s"].shift(-5)
df
```

```python
df["d"] = df["s"] - df["s5"]
value_list = df["d"]
def our_gate(x):
  if x>0:
    return 1
  else:
    return 0

result_list =[]
for i in value_list:
  y = our_gate(i)
  result_list.append(y)

df["p"] = result_list
df
```

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def sum_of_digits(n):
    return sum(int(d) for d in str(n))

n = 5001
while True:
    if is_prime(n) and is_prime(sum_of_digits(n)):
        print("The first prime number greater than 5000 whose digit sum is also prime is:", n)
        break
    n += 1
```

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False
    for i in range(3, int(n**0.5) + 1, 2):
        if n % i == 0:
            return False
    return True

def find_non_prime_interval(length):
    n = 1
    while True:
        if all(not is_prime(n + i) for i in range(length)):
            return n, n + length - 1
        n += 1

start, end = find_non_prime_interval(73)
print(f"Interval of 73 consecutive non-prime numbers found: from {start} to {end}")
```
